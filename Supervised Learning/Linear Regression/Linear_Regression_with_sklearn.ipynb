# Data Scaling & StandardScaler — Key Doubts Clarified

## 1. Why do we standardise or normalise data?

Raw feature scales distort learning.

### Problems without scaling
- Large-scale features dominate small-scale ones
- Gradient-based optimisation becomes slow or unstable
- Regularisation penalises features unfairly
- Numerical instability (ill-conditioned matrices)
- Coefficients become incomparable

### Algorithms that need scaling
- Linear / Logistic Regression
- SVM
- k-NN
- Neural Networks
- PCA

### Algorithms that do NOT need scaling
- Decision Trees
- Random Forest
- Gradient Boosting

Reason: tree splits depend on order, not magnitude.

---

## 2. Standardisation vs Normalisation

### Standardisation (Z-score)
\[
x' = \frac{x - \mu}{\sigma}
\]

- Mean = 0
- Standard deviation = 1
- Used for most ML algorithms

### Normalisation (Min–Max)
\[
x' = \frac{x - x_{min}}{x_{max} - x_{min}}
\]

- Fixed range \([0,1]\)
- Used when bounds matter

---

## 3. Why `fit_transform` on `X_train` and only `transform` on `X_test`?

To **avoid data leakage**.

- Training data defines the rules
- Test data only follows those rules

If `X_test` is used in `fit`:
- Test information leaks into training
- Evaluation becomes unrealistic
- Real-world performance is misrepresented

---

## 4. What does `fit_transform` actually do?

- `fit` → computes **mean (μ)** and **standard deviation (σ)**
- `transform` → applies scaling using those values

So:
- `fit_transform(X_train)`  
  → computes μ, σ + scales data
- `transform(X_test)`  
  → uses **same μ, σ**, no recomputation

---
